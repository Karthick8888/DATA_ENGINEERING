```python
def fibo(i):
    if i==1:
        return 1
    return i*fibo(i-1)
for i in range(1,20):
    print(fibo(i))
```

```python
x=[1,2,3,11,24,59,61]
def prime(i):
    if((i<=3) and (i>0)):
        return True
    for d in range(4,int(i/2)):
        if i%d==0:
            return False
    return True
for data in x:
    if prime(data):
        print(data)
```
