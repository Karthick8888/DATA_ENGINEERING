```python

def add(a, b):
    """Return the sum of two numbers."""
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Both arguments must be numbers")
    return a + b
def sub(a, b):
    """Return the sum of two numbers."""
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Both arguments must be numbers")
    return a - b

def mul(a, b):
    """Return the sum of two numbers."""
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Both arguments must be numbers")
    return a * b

def divide(a, b):
    """Return the division of a by b."""
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Both arguments must be numbers")
    if b == 0:
        raise ValueError("Division by zero is not allowed")
    return a / b
def modulo(a, b):
    """Return the division of a by b."""
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Both arguments must be numbers")
    if b == 0:
        raise ValueError("Division by zero is not allowed")
    return a % b

def factorial(n):
    if n<0:
        return -1
    if n==0:
        return 0
    if n==1:
        return 1
    return n*factorial(n-1)
```

```python
import unittest  #module

class TestMyFunctions(unittest.TestCase):
    
    def test_add_positive_numbers(self):
        self.assertEqual(add(2, 3), 5)

    def test_factorial(self):
        self.assertEqual(factorial(5),120)

    def test_factorial_negative_number(self):
        self.assertEqual(factorial(-5),-1)

    def test_factorial_zero_number(self):
        self.assertEqual(factorial(0),0)

    def test_add_negative_numbers(self):
        self.assertEqual(add(-2, -3), -5)

    def test_add_invalid_type(self):
        with self.assertRaises(TypeError):
            add(2, "three")

    def test_divide_normal(self):
        self.assertAlmostEqual(divide(10, 2), 5.0)

    def test_divide_float(self):
        self.assertAlmostEqual(divide(10.6, 2), 5.3)

    def test_sub_negative_numbers(self):
        self.assertEqual(sub(-2, -3), 1)

    def test_add_invalid_type(self):
        with self.assertRaises(TypeError):
           add(2, "three")

    def test_divide_by_zero(self):
        with self.assertRaises(ValueError):
            divide(5, 0)
    def test_modulo_normal(self):
        self.assertAlmostEqual(modulo(10, 2), 0)

    def test_modulo_float(self):
        self.assertAlmostEqual(modulo(10.6, 2),0.6)

    def test_modulo_by_zero(self):
        with self.assertRaises(ValueError):
            modulo(5, 0)


unittest.main(argv=[''],verbosity=2,  exit=False)
```
