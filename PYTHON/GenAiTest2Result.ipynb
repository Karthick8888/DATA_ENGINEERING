```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'Student_sheet_Output - Sheet1.csv')
df


```

```python
df.describe()
```

```python


# Calculate total score per person
df["total_score"] = df.loc[:, "Question1":"Question27"].sum(axis=1)

# Calculate percentage score
df["percentage"] = df["total_score"] / 28 * 100

# Item analysis: how many people got each question correct
item_analysis = df.loc[:, "Question1":"Question27"].sum(axis=0)

print(item_analysis)
```

```python


# Calculate total score for each student
df["total_score"] = df.loc[:, "Question1":"Question27"].sum(axis=1)

# Preview results
print(df[["NAME", "total_score"]])
```

```python
df["percentage"] = (df["total_score"] / 28) * 100

print(df[["NAME", "total_score", "percentage"]])
```

```python
# Classify students based on percentage 
def classify(pct):
    if pct >= 60: 
        return "Advanced"
    elif pct >= 50: 
        return "Upper Intermediate"
    elif pct >= 40: 
        return "Intermediate"
    else: 
        return "Beginner" 
df["level"] = df["percentage"].apply(classify) 
# Preview results
print(df[["NAME", "total_score", "percentage", "level"]])
```

```python
df.to_csv("student_scores.csv", index=False)

print("Excel file 'student_scores.xlsx' has been created successfully!")
```
