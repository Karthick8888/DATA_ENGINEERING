```python
#mapreduce()

num = [1,2,3,4]

#case 1
def hallsquare(num):
    return (num**2)

#using map() function

#case 2
squared = list(map(hallsquare,num))

print(squared)

#case 3

squared = list(map(lambda x:x**2,num))
print(squared)
```
