```python
gen_over = [i for i in range(1,21)]

india_runrate = [10,13,25,8,11,21,22,14,10,9,8,17,6,17,13,10,12,13,11,14]
SA_runrate = [9,13,7,8,11,8,12,13,14,9,12,15,8,19,12,11,14,8,8,8]
india_score = np.linspace(10,sum(india_runrate),20)
SA_score = np.linspace(10,sum(SA_runrate),20)

plt.plot(india_score,india_runrate,marker='o',linestyle='-',color= 'blue',label='TEAM A')
plt.plot(SA_score,SA_runrate,marker='^',linestyle='-.',color= 'green',label='TEAM B')

plt.xlabel('Run Per Over')
plt.ylabel('Score')
plt.title('RUN RATE')
plt.show()

print("INDIA : ",sum(india_runrate))
print("SA : ",sum(SA_runrate))
if sum(india_runrate) > sum(SA_runrate):
    win = "INDIA"
    win_run = sum(india_runrate) - sum(SA_runrate)
else:
    win = "SA"
    win_run = sum(SA_runrate)-sum(india_runrate)
print("{0} WON BY {1} RUNS".format(win,win_run))
```
