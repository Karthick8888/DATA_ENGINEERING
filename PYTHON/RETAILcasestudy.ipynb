```python
grouped = ds.groupby(['store_location','product_category'])['quantity'].sum().reset_index()
idx = grouped.groupby('store_location')['quantity'].idxmax()
top_selling = grouped.loc[idx]
top_selling
```

# MOTIVE:
# TO UNDERSTAND PRODUCT DEMAND ACROSS DIFF LOCATIONS

```python
pivot = ds.groupby(['store_location','product_category'])['quantity'].sum().unstack()
pivot.plot(kind='bar')
plt.ylabel("Total Quantity Sold")
plt.title("Top Selling Products by Location")
plt.show()
```

```python
pivot = ds.groupby(['store_location','customer_type'])
pivot.plot(kind='bar')
plt.ylabel("Total Quantity Sold")
plt.title("Top Selling Products by Location")
plt.show()
```
