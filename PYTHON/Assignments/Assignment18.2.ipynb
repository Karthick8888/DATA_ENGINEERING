1.Matrix transpose using with and without built in function

```python
#without built-in
matrix=[[1,2,3],[4,5,6],[7,8,9]]
transpose=[]
for i in range(len(matrix[0])):
    row=[]
    for j in range(len(matrix)):
        row.append(matrix[j][i])
    transpose.append(row)
print(transpose)
```

```python
#with built-in
matrix=[[1,2,3],[4,5,6],[7,8,9]]
transpose=list(zip(*matrix))
print(transpose)
```

2.Assignment share elements without sharing address

```python
#copy()
a=[1,2,3]
b=a.copy()
print(a)
print(id(a))
print(b)
print(id(b))
```

```python
#slice
a=[1,2,3]
b=a[:]
print(a)
print(id(a))
print(b)
print(id(b))
```

3.How to reassign the data that is temporarily deleted

```python
a=[1,2,3]
b=a
del a
print(b)  
```
